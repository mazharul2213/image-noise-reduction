```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/37.png', 0)

#blurx = cv2.blur(img, ksize=(1,1))
# global thresholding
ret1, th1 = cv2.threshold(img,140,255,cv2.THRESH_BINARY)

# Otsu's thresholding
#ret2, th2 = cv2.threshold(img, 130,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Otsu's thresholding after Gaussian filtering
blur1 = cv2.medianBlur(img, 1)
blur2 = cv2.GaussianBlur(img,(1, 1), 3)
blur3 = cv2.blur(img, ksize=(1,1))
blur4 = cv2.bilateralFilter(img, 3, sigmaSpace = 30, sigmaColor = 30)

_,th3 = cv2.threshold(blur2, 120, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

images = [ th1, th3]

fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (40, 40))
for ind, p in enumerate(images):
    ax = axs[ind]
    ax.imshow(p)
plt.show()


#cv2.imwrite("done/img1.jpg", th1)
cv2.imwrite("done/img2.jpg", th3)
```
